```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_image(image_path):
    """Carica un'immagine TIFF/JPG/PNG e la converte in RGB."""
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    if img is None:
        raise ValueError(f"Impossibile caricare: {image_path}")
    if len(img.shape) == 2:  # Immagine monocromatica
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    else:  # Immagine a colori (BGR -> RGB)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def read_yolo_polygons(txt_path, img_width, img_height):
    """Legge un file YOLO con poligoni e restituisce liste di coordinate assolute."""
    polygons = []
    if not os.path.exists(txt_path):
        return polygons
    
    with open(txt_path, 'r') as file:
        for line in file:
            parts = list(map(float, line.strip().split()))
            if len(parts) < 3 or len(parts) % 2 != 1:  # Deve avere class_id + coppie x/y
                continue
            class_id = int(parts[0])
            points = np.array(parts[1:]).reshape(-1, 2)  # Converti in [[x1,y1], [x2,y2], ...]
            points[:, 0] *= img_width  # Converti x in pixel
            points[:, 1] *= img_height  # Converti y in pixel
            polygons.append((class_id, points.astype(int)))
    return polygons

def draw_polygons(image, polygons, class_names=None):
    """Disegna poligoni sull'immagine."""
    colors = plt.cm.get_cmap('tab10', len(class_names) if class_names else 10)
    for class_id, points in polygons:
        color = colors(class_id % 10)[:3]
        color = tuple(int(255 * c) for c in color)
        # Disegna il poligono
        cv2.polylines(image, [points], isClosed=True, color=color, thickness=2)
        # Aggiungi etichetta
        label = class_names[class_id] if class_names else str(class_id)
        cv2.putText(image, label, tuple(points[0]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
    return image

def show_single_image(image_path, label_path=None, class_names=None):
    """Mostra un'immagine con i poligoni (se presenti)."""
    image = load_image(image_path)
    img_height, img_width = image.shape[:2]
    
    if label_path and os.path.exists(label_path):
        polygons = read_yolo_polygons(label_path, img_width, img_height)
        image = draw_polygons(image, polygons, class_names)

    plt.figure(figsize = (15,15))
    plt.imshow(image)
    plt.axis('off')
    plt.title(os.path.basename(image_path))
    plt.show()

```
